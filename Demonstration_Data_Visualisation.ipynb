```python
import pandas as pd

df = pd.read_csv('demoStatsGrid.csv')
df.drop(index=df.index[0], axis=0, inplace=True)
display(df)
```

```python
df.describe()
```

```python
df[' Total Steps'].sum()
```

```python
# @title  Total Reward

from matplotlib import pyplot as plt
df[' Total Reward'].plot(kind='line', figsize=(8, 4), title=' Grid World Demo Total Reward')
plt.ylim((0,1))
plt.xlim((0))
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.axhline(y=0.664338, color='r', linestyle='dotted')
plt.legend(["reward", "mean reward"], loc="lower right")

plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
df = pd.read_csv('demoStatsExit.csv')
df.drop(index=df.index[0], axis=0, inplace=True)


df = df.drop(df[df[" Total Reward"] < 0.62].index)

from matplotlib import pyplot as plt
df[' Total Reward'].plot(kind='line', figsize=(8, 4), title=' Find Exit Agent Demo Total Reward')
plt.ylim((0,2))
plt.xlim((0))
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.axhline(y=1.444716, color='r', linestyle='dotted')
plt.legend(["reward", "mean reward"], loc="lower right")

plt.gca().spines[['top', 'right']].set_visible(False)
```
